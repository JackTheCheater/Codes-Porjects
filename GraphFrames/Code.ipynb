## Exercise 3: querying (30 min)

We want to find out what are **the most common destinations in the dataset from location to location**. We can approach in the following way:
- We can do this by performing a grouping operator and adding the edge counts together. 
- We obtain a new graph except each edge will now be the sum of all of the semantically same edges. 
- To get the result, we have to order it.

````python
# compute the top ten most common trips: complete here
topTrips = stationGraph.edges.groupBy("src", "dst").count().orderBy("count",ascending=False ).limit(10)
````

````python
topTrips.show()
````

In this instance, we have a directed graph. This means that our trips are directional from one location to another one. 
We therefore can find **the number of trips which go into a specific station and leave from a specific station**. Naturally, we can sort this information and we can find the stations with the highest inbound and outbound trips using the notion of indegree and outdegree.

````python
# Plots in / out count from stations

def plot_degrees(inDeg, outDeg):
    import matplotlib.pyplot as plt
    import warnings
    warnings.filterwarnings("ignore")
    inDegPd = inDeg.toPandas().sort_values('id')
    outDegPd = outDeg.toPandas().sort_values('id')
    fig, (ax1, ax2) = plt.subplots(1, 2, sharey = True )
    fig.set_size_inches(20, 5)
    fig.suptitle('in / out trips count by stations')
    ax1.set_xticklabels(inDegPd["id"].tolist())
    ax1.bar(inDegPd["id"],inDegPd["inDegree"],color='g' )
    ax1.set_title("top incoming trips")
    ax2.set_xticklabels(outDegPd["id"].tolist())
    ax2.bar(outDegPd["id"],outDegPd["outDegree"],color='b' )
    ax2.set_title("top outcoming trips")
    ax1.set_ylabel("count")
    fig.autofmt_xdate(bottom=0.2, rotation=30, ha='right')
    fig.show()
````

````python
# calculate indegree and outdegree: complete here
inDeg = stationGraph.inDegrees
outDeg = stationGraph.outDegrees

# plot
plot_degrees(inDeg.orderBy("inDegree", ascending=False).limit(20), outDeg.orderBy("outDegree", ascending=False).limit(20))
````

One additional interesting question on which we can focus is what is the station with the highest ratio of indegree but the lowest of outdegree: a station where trips end at but rarely start from. 
- Hint 1 : use the ratio between in/out for all station.
- Hint 2 : use pySparkSql *join* between *inDeg* and *outDeg* on column *id*.

````python
# Plots 2 colums histogram

def plot_hist(sparkData, arg, title, ylabel ) :
    import matplotlib.pyplot as plt
    import warnings
    warnings.filterwarnings("ignore")
    data = sparkData.toPandas()
    fig, ax1 = plt.subplots(1, 1, sharey = True )
    fig.set_size_inches(25, 10)
    fig.suptitle(title)
    ax1.set_xticklabels(data["id"].tolist())
    ax1.bar(data["id"],data[arg],color='r' )
    ax1.set_ylabel(ylabel)
    fig.autofmt_xdate(bottom=0.2, rotation=30, ha='right')
    fig.show()
````

````python
# compute the stations with highest indegree/outdegree ration: complete here
degreeRatio = inDeg.join(outDeg, inDeg["id"] == outDeg["id"]).drop(outDeg["id"]).selectExpr("id", "double(inDegree)/double(outDegree) as degreeRatio")
````

````python
# rank the rations by sorting for degree ratio: complete here
degreeRatio.orderBy("degreeRatio", ascending=False).show()
````

````python
plot_hist(degreeRatio.orderBy("degreeRatio", ascending=False), arg = "degreeRatio", title = "in/out ratio", ylabel = "ratio" )
````

As we can see in the histogram, the highest ratio correspond to the **Redwood City Medical Center**, and the second is the **Redwood City Library** (and it makes totally sense).

## Exercise 4: graph algorithms (30 min)

### PageRank
One of the most interesting algorithm we can apply to graphs is google's page rank and find the **most connected station**.

````python
from pyspark.sql.functions import desc
# find most connected station: complete here using pageRank()
ranks = stationGraph.pageRank(resetProbability=0.15, maxIter=10)
result = ranks.vertices.orderBy(desc("pagerank")).select("id", "pagerank")
plot_hist(result, arg = "pagerank", title = "PageRank!", ylabel = "rank" )
````

### BFS
Breadth-first Search will search our graph for how to connect two given nodes based on the edges in the graph. In
our context, we may want to do this to find the shortest paths to different stations. We can specify the maximum of
edges to follow with the *maxPathLength* and we can also specify an *edgeFilter* to filter out certain edges that
do not meet a certain requirement, like trips during non-business hours. Well choose two fairly close stations so that this does not run too long. However, you can do some pretty interesting graph traversals when you have some sparse graphs that have some distant connections. Feel free to play around with the stations (especially those in other cities) to see if you can get some distant stations to connect.

````python
# compute BFS: complete here using bfs()
stationGraph.bfs(fromExpr="id = 'Townsend at 7th'",  toExpr="id = 'Spear at Folsom'", maxPathLength=3).show(20)
````

## Exercise 5: motif (40 min)
Motifs are a way of expressing structural patterns in a graph. When we specify a motif, we are looking for patterns in the data instead of actual data. Our current dataset does not suit this sort of querying because our graph consists of
individual trips, not repeated interactions of certain individuals or identifiers. In GraphFrames, we specify our query in a domain-specific language. We specify combinations of vertices and edges. For example, if we want to specify a given vertex connects to another vertex we would specify `(a)-[ab]->(b)`. 
The letters inside of parenthesis or brackets do not signify values, but signify what the columns should be named in the resulting dataframe. We can omit the names (e.g.,` (a)-[]->()`) if we do not intend to query the resulting values.
Lets perform a query. In plain english, lets find all the round trip rides with two stations in between. We express this with the following motif using the find method to query our GraphFrame for that pattern. `(a)` signifies the starting station, `[ab]` represents an edge from `(a)` to our next station `(b)`. We repeat this for stations `(b)` to `(c)` and then from `(c)` to `(a)`.

````python
# compute the requested motif: complete here using find()
# sampling the data for faster computation of proof of concept
subgraph = GraphFrame(stationVertices, tripEdges.sample(False, 0.2))

#complete here
motifs = subgraph.find("(a)-[ab]->(b); (b)-[bc]->(c); (c)-[ca]->(a)")
````

The resulting dataframe contains nested fields for vertices a, b, and c as well as the respective edges. Now we can query that to answer a specific question. Given a certain bike:  
- **What is the shortest round trip time where that bike is taken from one station `(a)`, ridden to another, dropped off
  at `(b)`, ridden to another, dropped off at `(c)`, and then ridden back to the original station, `(a)`?**  

The following logic will parse our timestamps into Spark timestamps and then well do comparisons to make sure that its the same bike, traveling from station to station, and that the start times for each trip are correct.

````python
from pyspark.sql.functions import expr
# complete here
res = motifs
res = motifs.where("ca.`bike_id` = bc.`bike_id`")
res = res.where("ab.`bike_id` = bc.`bike_id`")
res = res.where("a.`id` != b.`id`")
res = res.where("b.`id` != c.`id`")
res = res.where("ab.`start_date` < bc.`start_date`")
res = res.where("bc.`start_date` < ca.`start_date`")
````

To answer the question, we may take the roundtrip and then we can order by (end_date - start_date) and take the first, in a map-reduce way.
