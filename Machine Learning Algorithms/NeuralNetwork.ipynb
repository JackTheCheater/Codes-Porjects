```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

<h4>Function used for plotting graphs of accuracy and loss

```python
def plotting_results(history):
    fig, ax=plt.subplots(2)
    ax[0].plot(history.history['accuracy'])
    ax[0].plot(history.history['val_accuracy'])
    ax[0].set_xlabel('Epoch')
    ax[0].set_ylabel('Accuracy')
    ax[0].set_title('Model accuracy')
    ax[0].legend(['Train','Test'], loc='lower right')
    ax[1].plot(history.history['loss'])
    ax[1].plot(history.history['val_loss'])
    ax[1].set_xlabel('Epoch')
    ax[1].set_ylabel('Loss')
    ax[1].set_title('Model loss')
    ax[1].legend(['Train','Test'], loc='upper right')
    fig.tight_layout()
    plt.show()
```

<h4>Data Loading - 10 dimensional vectors

```python
rawdata=np.loadtxt('semeion.data');
x=rawdata[:,0:256].copy();
t=rawdata[:,-10:rawdata.shape[1]].copy();
del(rawdata);
x_train,x_test,t_train,t_test=train_test_split(x,t,test_size=0.33)
```

<h4>First Model : input layer, hidden layer (32 units), output layer (10 units)

```python
#Model definition
model=tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(32, input_dim=256, activation='sigmoid'))
model.add(tf.keras.layers.Dense(10, activation='softmax'))

#Model compilation
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

#Fitting the model on the dataset (set verbose=2 if you want to see the training progress)
history=model.fit(x_train, t_train, validation_data=(x_test,t_test), shuffle=True, epochs=100, batch_size=64, verbose=0)

#Evaluation of keras model with accuracy score
loss, accuracy=model.evaluate(x_test, t_test, verbose=0)
print('Accuracy: %2.f'%(accuracy*100))
```

```python
plotting_results(history)
```

```python
plotting_results(history)
```
